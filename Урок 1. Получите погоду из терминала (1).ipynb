```python
import requests
```

```python
url = 'https://wttr.in'
```

```python
def get_weather(countries):
    params = {
        'lang' : 'ru',
        'M':'',
        'm':'',
        'n':'',
        'q':'',
        'T':''
    }
    for place in countries:
        response = requests.get(f'{url}/{place}', params = params)
        response.raise_for_status()
        print(response.text)
    return
```

```python
def main():
  countries = ['london', 'svo', 'череповец']
  get_weather(countries)
```

```python
if __name__ == '__main__':
    main()
```
